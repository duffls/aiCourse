```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

iris = load_iris()
iris_data = iris.data
iris_data_pd = pd.DataFrame(iris_data, columns=iris.feature_names)
iris_data_pd
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.5, min_samples=10)
db.fit(iris_data_pd.iloc[:, 2:4])

y_pred = db.fit_predict(iris_data_pd.iloc[:, 2:4])

plt.scatter(iris_data_pd.iloc[:, 2], iris_data_pd.iloc[:, 3], c=y_pred)
plt.title('Density Clustering')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

```python
db.labels_
```

```python
db.set_params(eps=0.5, min_samples=5)
y_pred = db.fit_predict(iris_data_pd.iloc[:, 2:4])
plt.scatter(iris_data_pd.iloc[:, 2], iris_data_pd.iloc[:, 3], c=y_pred)
plt.title('Density Clustering')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
print(db.labels_)
```

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False # - ë¬¸ìž ê¹¨ì§€ëŠ” ë¶€ë¶„ í•´ê²°
x, y = make_moons(n_samples=200, noise=0.05, random_state=0) # dataset
# MinMaxScaler ë©”ì†Œë“œë¡œ ì „ì²˜ë¦¬
scaler_MMS = MinMaxScaler().fit(x) 
x_scaled_MMS = scaler_MMS.transform(x) # ì „ì²˜ë¦¬ ë©”ì†Œë“œë¥¼ í›ˆë ¨ë°ì´í„°ì— ì ìš©
dbscan = DBSCAN() # ëª¨ë¸ìƒì„±
clusters_MMS = dbscan.fit_predict(x_scaled_MMS) # ëª¨ë¸ í•™ìŠµ
print('np.unique(clusters_MMS) ì˜ˆì¸¡í•œ ë ˆì´ë¸”: {}'.format(np.unique(clusters_MMS))) # [0]

df = np.hstack([x_scaled_MMS, clusters_MMS.reshape(-1, 1)]) # x_scaled_ss ì˜¤ë¥¸ìª½ì— 1ì—´ ë¶™ì´ê¸°
df_ft0 = df[df[:,2]==0, :] # í´ëŸ¬ìŠ¤í„° 0 ì¶”ì¶œ
df_ft1 = df[df[:,2]==1, :] # í´ëŸ¬ìŠ¤í„° 1 ì¶”ì¶œ

plt.scatter(df_ft0[:, 0], df_ft0[:, 1], label='cluster 0', cmap='Pairs') # x, y, label, ìƒ‰ìƒ
plt.scatter(df_ft1[:, 0], df_ft1[:, 1], label='cluster 1', cmap='Pairs')
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.legend()
plt.show()
```

```python
# StandardScaler ë©”ì†Œë“œë¡œ ì „ì²˜ë¦¬
scaler_ss = StandardScaler().fit(x)
x_scaled_ss = scaler_ss.transform(x) 
dbscan = DBSCAN()
clusters_ss = dbscan.fit_predict(x_scaled_ss)
print('np.unique(clusters_ss) ì˜ˆì¸¡í•œ ë ˆì´ë¸”:{}'.format(np.unique(clusters_ss))) # [0 ,1]

df = np.hstack([x_scaled_ss, clusters_ss.reshape(-1, 1)]) # x_scaled_ss ì˜¤ë¥¸ìª½ì— 1ì—´ ë¶™ì´ê¸°
df_ft0 = df[df[:,2]==0, :] # í´ëŸ¬ìŠ¤í„° 0 ì¶”ì¶œ
df_ft1 = df[df[:,2]==1, :] # í´ëŸ¬ìŠ¤í„° 1 ì¶”ì¶œ

# matplotlibë¡œ ê·¸ëž˜í”„ ê·¸ë¦¬ê¸°
plt.scatter(df_ft0[:, 0], df_ft0[:, 1], label='cluster 0', cmap='Pairs') # x, y, label, ìƒ‰ìƒ
plt.scatter(df_ft1[:, 0], df_ft1[:, 1], label='cluster 1', cmap='Pairs')
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt.legend()
plt.show()
```

```python
# StandardScalerë§Œ ì‚¬ìš©í•œ ì˜ˆì œ
from sklearn.datasets import make_moons # two_moonsë°ì´í„°ì…‹ ìƒì„±
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import mglearn
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False # - ë¬¸ìž ê¹¨ì§€ëŠ” ë¶€ë¶„ í•´ê²°
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
dbscan = DBSCAN() # ê¸°ë³¸ê°’ eps=0.5, eps=0.7, eps=0.2
clusters = dbscan.fit_predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap=mglearn.cm2, s=60, edgecolors='black')
plt.xlabel("feature 0")
plt.ylabel("feature 1")
plt.show()
```

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
import numpy as np

people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
```

```python
print(people.keys())
print(people.data.shape)
print(people.images.shape)
print(people.target)
print(people.target_names)
print(people.DESCR)
```

```python
x, cnt = np.unique(people.target, return_counts=True)
print(x)
print(cnt)
```

```python
image_shape = people.images[0].shape
mask = np.zeros(people.target.shape, dtype=np.bool)

mask
```

```python
# ìµœëŒ€ 50ìž¥ ê¹Œì§€ë§Œ ê°€ì ¸ì˜¤ê²Œ ì œí•œ.
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1
    #print(np.where(people.target == target))
mask
```

```python
X_people = people.data[mask]
y_people = people.target[mask]

print(X_people.shape)
print(y_people.shape)

x, cnt = np.unique(y_people, return_counts=True)
print(x)
print(cnt)
```

```python
# 0~255 ì‚¬ì´ì˜ í‘ë°± ì´ë¯¸ì§€ë¥¼ í”½ì…€ ê°’ì„ 0~1 ìŠ¤ì¼€ì¼ë¡œ ì¡°ì •
# MinMaxScaler ì ìš©ê³¼ ë¹„ìŠ·
X_people = X_people / 255.
# ì°¨ì› ì¶•ì†Œ(PCA__)
pca = PCA(n_components=100, whiten=True, random_state=0)
pca.fit(X_people)
X_pca = pca.transform(X_people)
X_pca.shape
```

# 1. DBSCANìœ¼ë¡œ ì–¼êµ´ ë°ì´í„°ì…‹ ë¶„ì„í•˜ê¸°

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN()
labels = dbscan.fit_predict(X_pca)
print("ê³ ìœ í•œ ë ˆì´ë¸” \n{}".format(np.unique(labels)))
```

```python
# min_samples=3 ì¸ ê²½ìš°
dbscan = DBSCAN(min_samples=3)
labels = dbscan.fit_predict(X_pca)
print("ê³ ìœ í•œ ë ˆì´ë¸” \n{}".format(np.unique(labels)))
```

```python
# min_samples=3, eps=15 ì¸ ê²½ìš°
dbscan = DBSCAN(min_samples=3, eps=15)
labels = dbscan.fit_predict(X_pca)
print("ê³ ìœ í•œ ë ˆì´ë¸” \n{}".format(np.unique(labels)))
print("í´ëŸ¬ìŠ¤í„° ë³„ í¬ì¸íŠ¸ ìˆ˜: {}".format(np.bincount(labels + 1))) # np.bincount : ìŒìˆ˜ëŠ” ì‚¬ìš©ë¶ˆê°€
```

```python
import matplotlib.pyplot as plt
print("í´ëŸ¬ìŠ¤í„° ë³„ í¬ì¸íŠ¸ ìˆ˜: {}".format(np.bincount(labels + 1)))
noise = X_people[labels == -1] # ìž¡ìŒ(noise) ì´ë¯¸ì§€ ì¶”ì¶œ
fig, axes = plt.subplots(4, 9, subplot_kw={'xticks':(), 'yticks':()}, figsize=(12, 4))
for image, ax in zip(noise, axes.ravel()): # ìž¡ìŒ(noise) ì´ë¯¸ì§€ ì¶œë ¥
    ax.imshow(image.reshape(image_shape), vmin=0, vmax=1)
```

```python
for eps in [1, 3, 5, 7, 9, 11, 13]:
    for min_samples in [3, 5, 7, 9]:
        print("\neps={}, min_samples={}".format(eps, min_samples)) 
        dbscan = DBSCAN(eps=eps, min_samples=min_samples) 
        labels = dbscan.fit_predict(X_pca) 
        print("í´ëŸ¬ìŠ¤í„° ê°’ {}".format(np.unique(labels))) 
        print("í´ëŸ¬ìŠ¤í„° ìˆ˜ {}".format(len(np.unique(labels))))
        print("í´ëŸ¬ìŠ¤í„° í¬ê¸° {}".format(np.bincount(labels + 1))) 
```

```python
# eps=7ì˜ 13ê°œì˜ í´ëŸ¬ìŠ¤í„°ì— ìžˆëŠ” ëª¨ë“  í¬ì¸íŠ¸ë¥¼ ì‹œê°í™”
dbscan = DBSCAN(min_samples=3,eps=7)
labels = dbscan.fit_predict(X_pca)
for cluster in range(max(labels) + 1): 
    mask = labels == cluster # labels == clusterì— ì˜í•´ labelsê°€ -1ì¸ ê²½ìš°ëŠ” ìŠ¤í‚µ
    n_images = np.sum(mask)
    fig, axes = plt.subplots(1, n_images,figsize=(n_images * 1.5, 4),subplot_kw={'xticks':(), 'yticks':()}) 
    for image, label, ax in zip(X_people[mask], y_people[mask], axes):
        ax.imshow(image.reshape(image_shape), vmin=0, vmax=1)
        ax.set_title(people.target_names[label].split()[-1])
```

# 2. k-í‰ê· ìœ¼ë¡œ ì–¼êµ´ ë°ì´í„°ì…‹ ë¶„ì„í•˜ê¸°

```python
from sklearn.cluster import KMeans
# k-í‰ê· ìœ¼ë¡œ í´ëŸ¬ìŠ¤í„° ì¶”ì¶œ
km = KMeans(n_clusters=13, random_state=0)
labels_km = km.fit_predict(X_pca)
print("k-í‰ê· ì˜ í´ëŸ¬ìŠ¤í„° í¬ê¸° : {}".format(np.bincount(labels_km)))
```

```python
# k-í‰ê·  í´ëŸ¬ìŠ¤í„° ì¤‘ì‹¬ ì‹œê°í™”
fig, axes = plt.subplots(2, 5, figsize=(12, 4), subplot_kw={'xticks':(), 'yticks':()}) 
for center, ax in zip(km.cluster_centers_, axes.ravel()):
    ax.imshow(pca.inverse_transform(center).reshape(image_shape), vmin=0, vmax=1)
```

```python
# ê° í´ëŸ¬ìŠ¤í„°ì—ì„œ ê°€ìž¥ ëŒ€í‘œë˜ëŠ” ì´ë¯¸ì§€ ë‹¤ì„¯ê°œì™€ ê°€ìž¥ ë¨¼ ì´ë¯¸ì§€ ë‹¤ì„¯ê°œ í‘œí˜„
import mglearn
mglearn.plots.plot_kmeans_faces(km, pca, X_pca, X_people, y_people, people.target_names)
```
