```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
dataset = load_iris()
print(dataset.keys())
print(dataset.DESCR)
```

```python
data = pd.DataFrame(dataset.data, columns=dataset.feature_names)
X_train, X_test, y_train, y_test = train_test_split(data, dataset.target, test_size=0.25, random_state=42)

y_train_scaled = y_train
```

```python
# 학습데이터 전처리 전처리(scale)
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train_scaled = mms.fit_transform(X_train)
X_test_scaled = mms.fit_transform(X_test)
print("X_train_scaled : \n", X_train_scaled)
print("X_test_scaled : \n", X_test_scaled)
```

# KNeighborsClassifier Parameters

- n_neighbors: 분류시 고려할 인접 샘플 수(가장 가까운 이웃을 몇개 고를지에 대한 param) defaule=5
- weights: 'distance'-인접한 샘플의 거리에 따른 다른 가중치 부여(가까울수록 큰 가중치) def='uniform'
- metric:거리계산의 척도(minkowski, euclidean, manhattan등),def='minkowski'."p"라는 매개변수를 추가사용. p=1 manhattan, p=2 euclidean과 동일. 
ex)KNeighborsClassifier(n_neighbors=11,p=2,metric='minkowski') p=2가 기본값
- n_jobs: 사용할 코어 수. -1이면 전부 사용

```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# 그래프 출력을 위해 선언합니다.
%matplotlib inline
# sklearn 모델의 동일핚 결과 출력을 위해 선언합니다.
np.random.seed(5)
```

```python
train = pd.read_csv("https://raw.githubusercontent.com/wikibook/machine-learning/2.0/data/csv/basketball_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/wikibook/machine-learning/2.0/data/csv/basketball_test.csv")
print(train)
print('--------------------------------------------')
print(test)
```

# 속성(컬럼) 설명
- Player: 이름
- Pos: 포지션
- 3P: 한 경기 평균 3점슛 성공횟수
- TRB: " 리바운드 
- BLK: " 블락

```python
# import kNN library
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
# find best k, range from 3 to half of the number of data
max_k_range = train.shape[0] // 2
max_k_range
```

```python
k_list = []
for i in range(3, max_k_range, 2):
    k_list.append(i)
k_list
```

```python
cross_validation_scores = []
x_train = train[['3P', 'BLK' , 'TRB']] # list를  DataFrame으로
y_train = train[['Pos']] #  ""

# 10-fold cross validation
for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x_train, y_train.values.ravel(), cv=10, scoring='accuracy')
    cross_validation_scores.append(scores.mean())
cross_validation_scores
```

```python
# visualize accuracy according to k
plt.plot(k_list, cross_validation_scores)
plt.xlabel('the number of k')
plt.ylabel('Accuracy')
plt.show()
```

```python
# find best k
cvs = cross_validation_scores
k = k_list[cvs.index(max(cross_validation_scores))]
print(f"The best number of k: {str(k)}")
```

```python
# import libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=k)

# select data features
x_train = train[['3P', 'BLK']]

# select target value
y_train = train[['Pos']]
```

```python
# setup knn using train data
#knn.fit(x_train, y_train.values.ravel())
knn.fit(x_train, y_train.values.ravel())
```

```python
# select data feature to be used for prediction
x_test = test[['3P', 'BLK']]

# select target value
y_test = test[['Pos']]
```

```python
# test
#pred = knn.predict(x_test)
pred = knn.predict(x_test)

pred
```

```python
comparison = pd.DataFrame({'prediction':pred, 'ground_truth':y_test.values.ravel()}) 
comparison
```

```python
# check accuracy
print("accuracy : " + str(accuracy_score(y_test.values.ravel(), pred)) )
```

```python
# 3개의 특징으로 예측하기 (3점슛, 블로킹, 리바운드)
knn = KNeighborsClassifier(n_neighbors=k)
# select data features to be used in train
x_train = train[['3P', 'BLK', 'TRB']]
# select target
y_train = train[['Pos']]
# y_train = train['Pos']  , y_test   (.values.ravel() 생략)
# build knn model
knn.fit(x_train, y_train.values.ravel())
# select features to be used for prediction
x_test = test[['3P', 'BLK', 'TRB']]
# select target
y_test = test[['Pos']]
# test
pred = knn.predict(x_test)
pred
```
